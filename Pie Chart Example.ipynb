```python
import matplotlib.pyplot as plt 
```

```python
#%matplotlib inline
```

```python
labels=['Male', 'Female', 'Trans']
sizes=[50,45,5]
```

```python
#explode
explodes=[0.1,0.1,0.2]
```

### Everything is good till this far, now we need to draw pie chart.

we will take the instance of subplots, that is ax

```python
fig, ax=plt.subplots()
ax.pie(sizes,
       labels=labels,
       explode=explodes,
       autopct='%1.1f%%',
       shadow=True, startangle=60) 
ax.set_title('Gender Distribution')

plt.show()
# ?fig
```
