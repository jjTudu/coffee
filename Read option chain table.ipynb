### Read csv option chain table NSE

```python
import pandas as pd
url='./tatasteeloc.csv'
rawData=pd.read_csv(url, header=1)
```

```python
rawData.info()
```

```python
rawData.columns
```

```python
rawData.columns=rawData.columns.str.replace(' ','_')
```

```python
rawData.head(3)
```

```python
rawData.replace(',','', regex=True, inplace=True)
df=rawData.apply(pd.to_numeric,errors='coerce')
df.dropna(axis=1, how='all', inplace=True)
```

```python
df.head()
```

```python
df.OI=((df.OI/df.OI.sum())*100).round(2)
pd.set_option('display.max_columns', None)
#movies.head()
df
```
